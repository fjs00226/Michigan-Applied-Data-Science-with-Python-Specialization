## Question 1
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NHL** using **2018** data.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

def nhl_correlation(): 
    # YOUR CODE HERE
    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    
    area_population=cities[['Metropolitan area','Population (2016 est.)[8]','NHL']]
    area_population.columns=['Metropolitan area','Population','NHL']
    area_population.replace('—|\[.*\]','',regex=True,inplace=True)
    
    # clean nhl
    nhl_2018=nhl_df[nhl_df.year==2018]
    #Win/Loss ratio refers to the number of wins over the number of wins plus the number of losses.
    nhl_2018=nhl_2018.drop([0,9,18,26])
    nhl_2018['wl%']=pd.to_numeric(nhl_2018['W'])/(pd.to_numeric(nhl_2018['W'])+pd.to_numeric(nhl_2018['L']))
    nhl_2018['wl%']=pd.to_numeric(nhl_2018['W'])/(pd.to_numeric(nhl_2018['W'])+pd.to_numeric(nhl_2018['L']))
    nhl_2018=nhl_2018[['team','wl%']]
    nhl_2018['Metropolitan area']=nhl_2018['team']
    nhl_2018['Metropolitan area']=nhl_2018['Metropolitan area'].str.split(' ',n=1,expand=True)[[0]]
    nhl_2018['Metropolitan area']=nhl_2018['Metropolitan area'].replace({'New':'New York City',
                                                                'Tampa':'Tampa Bay Area',
                                                                'Minnesota':'Minneapolis–Saint Paul',
                                                                'Kansas':'Kansas City',
                                                                'Los':'Los Angeles',
                                                                'Texas':'Dallas–Fort Worth',
                                                                'Washington':'Washington, D.C.',
                                                                'Miami':'Miami–Fort Lauderdale',
                                                                'St.':'St. Louis',
                                                                'Arizona':'Phoenix',
                                                                'Oakland':'San Francisco Bay Area',
                                                                'Colorado':'Denver',
                                                                'Anaheim':'Los Angeles',
                                                                'Carolina':'Raleigh',
                                                                'Dallas':'Dallas–Fort Worth',
                                                                'Vegas':'Las Vegas',
                                                                'Florida':'Miami–Fort Lauderdale',
                                                                'San':'San Francisco Bay Area'})
    
    One=area_population.merge(nhl_2018, how='inner',on='Metropolitan area').sort_values(by='Metropolitan area')
    One['Population']=pd.to_numeric(One['Population']) 
    nhl_wl = One.groupby('Metropolitan area',sort=True).mean()    
    
    #raise NotImplementedError()
    
    population_by_region = nhl_wl['Population'] # pass in metropolitan area population from cities
    win_loss_by_region = nhl_wl['wl%'] # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```
